# Deep Learning with PyTorch

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

## $XOR$
**Fitting the $XOR$ function (again!)**

Define the $XOR$ function & generate training data

```python
# Define the XOR function
def xor(x, y):
    return (x+y)%2

# Generate training data
X = [[0, 0], [0, 1], [1, 0], [1, 1]]
Y = [[xor(x, y)] for x, y in X]
```

Convert training data to tensors

```python
X_tensor = torch.Tensor(X)
Y_tensor = torch.Tensor(Y)
```

Define the neural network using the nn.Module class with:
- two fully connected layers
- relu activation for the hidden layer; no activation for the output layer
- the number of hidden units as a parameter, set to $6$ by default

```python
class Net(nn.Module):
    def __init__(self, hidden_units=6):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2, hidden_units)
        self.fc2 = nn.Linear(hidden_units, 1)

    def forward(self, x):
        x = nn.ReLU() ( self.fc1(x) )
        x = self.fc2(x)
        return x
```

Create an instance of the neural network and test its outputs (should be random noise)

```python
torch.manual_seed(42)

# create the network
net = Net()

# test
with torch.no_grad():
    print(net(X_tensor))
    plt.imshow( net(X_tensor).reshape((2,2)) )
    plt.colorbar()
    plt.show()
```

Define the loss function (MSE) and optimizer (SGD) with learning rate $0.01$

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(net.parameters(), lr=0.01)
```

Train the neural network for $5000$ epochs;
in addition:
- print the loss every $500$ epochs
- save the losses and plot them in a semilogy

```python
n_epochs = 5000
losses = np.zeros(n_epochs)
for epoch in range(n_epochs):
    # Forward pass
    outputs = net(X_tensor)
    loss = criterion(outputs, Y_tensor)
    # save & print the loss
    losses[epoch] = loss.detach().numpy()
    if epoch%500 == 0:
        print(f"Epoch {epoch}: loss = {loss}")
    # Backward pass and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
# semilogy of the loss
plt.semilogy(losses)
plt.show()
```

Test your model & plot its outputs

```python
with torch.no_grad():
  print(net(X_tensor))
  plt.imshow( net(X_tensor).reshape((2,2)) )
  plt.colorbar()
  plt.show()
```

## $\sin(x)$
**Fitting $\sin(x)$ on $[-2\pi, 2\pi]$**

Define the function we want to fit

```python
def target_function(x):
    return np.sin(x)
```

Define the bounds of the interval

```python
lower_bound, higher_bound = -2*np.pi, 2*np.pi
```

Generate some training data: $1000$ equispaced points on our interval.

```python
x_train = torch.linspace(lower_bound, higher_bound, 1000)[:,None]
y_train = target_function(x_train)
```

Define the neural network with:
- two dense layers
- $ReLU$ activation for the hidden layer (no activation for the output)
- sizes of input, hidden and output layers as parameters (default to $1$, $10$, $1$)

```python
class Net(nn.Module):
    def __init__(self, input_size=1, hidden_size=10, output_size=1):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.fc1(x)
        x = nn.ReLU()(x)
        x = self.fc2(x)
        return x
```

Create the nework to fit the $\sin$ function with a hidden layer of $100$ neurons; plot its current output on the interval

```python
torch.manual_seed(42)

# create the network
net = Net(1, 100, 1)

# plot current state
with torch.no_grad():
    plt.plot(x_train, y_train)
    plt.plot(x_train, net(x_train).detach())
    plt.show()
```

Define the loss function (MSE) and optimizer (SGD) with learning rate of $0.005$

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(net.parameters(), lr=0.005)
```

Train the network on $25000$ epochs printing the loss every $2500$ epochs & plot the loss in a semilogy

```python
n_epochs = 25000
losses = np.zeros(n_epochs)
for epoch in range(n_epochs):
    # Forward pass
    outputs = net(x_train)
    loss = criterion(outputs, y_train)
    losses[epoch] = loss
    # Backward and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    # print the loss
    if epoch % 5000 == 0:
        print(f'Epoch [{epoch}/{n_epochs}], Loss: {loss.item():.4f}')
        # # plot current state
        # x_test = torch.linspace(lower_bound, higher_bound, 90)[:,None]
        # plt.plot(x_train, y_train)
        # plt.plot(x_test, net(x_test).detach())
        # plt.show()
        # optimizer.zero_grad()

print('Loss (semilogy):')
plt.semilogy(losses)
plt.show()
```
