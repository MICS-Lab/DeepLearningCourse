```python
import torch
import torch.nn as nn
```

Define the network

```python
class SimpleNet(nn.Module):
    def __init__(self, input_size, hidden_layers, hidden_size, output_size):
        super(SimpleNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.list_hidden_layers = []
        for i in range(hidden_layers):
            self.list_hidden_layers.append(nn.Linear(hidden_size, hidden_size))
        self.fc3 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

Initialize the network

```python
input_size = 1_000
hidden_size = 5
output_size = 10
```

```python
net = SimpleNet(input_size, hidden_size, output_size)
```

Define a loss function and optimization algorithm


```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(net.parameters(), lr=1e-2)
```

Create a random input tensor

```python
inputs = torch.randn(1, input_size)
```

Train the network for multiple epochs

```python
for epoch in range(1_000):
    # Forward pass
    outputs = net(inputs)
    loss = criterion(outputs, torch.zeros(1, output_size))

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Check the gradient values of the last layer after each epoch
    print(f"Epoch {epoch}: gradient = {net.fc3.weight.grad.norm()}, loss = {loss.item()}")
```
